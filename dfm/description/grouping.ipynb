```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_pickle('csv/new.pkl')
df.head()
```

```python
mapping_dict = {
    "retsinformationdk": "Legal",
    "skat": "Legal",
    "retspraksis": "Legal",
    "hest": "Social Media",
    "cc": "Web",
    "adl": "Wiki & Books",
    "botxt": "Other",
    "danavis": "News",
    "dannet": "dannet",
    "depbank": "Other",
    "ep": "Conversation",
    "ft": "Conversation",
    "gutenberg": "Wiki & Books",
    "jvj": "Wiki & Books",
    "naat": "Conversation",
    "opensub": "Conversation",
    "relig": "Wiki & Books",
    "spont": "Conversation",
    "synne": "Other",
    "tv2r": "News",
    "wiki": "Wiki & Books",
    "wikibooks": "Wiki & Books",
    "wikisource": "Wiki & Books",
}
```

```python
df['source_mapped'] = df['source'].apply(lambda x: mapping_dict[x])
df.to_pickle('csv/new.pkl')
df.head()
```

```python
tokens = df[['tokens', 'source']].copy()
tokens.head()
```

```python
df0 = df.drop(['tokens'], axis=1)
```

```python
df0.head()
```

```python
df0['sum'] = df0.sum(axis=1)

df0.head()
```

```python
df0['tokens'] =  df['tokens']
```

```python
df0.head()
```

```python
ids = []  

for i in range(len(df0)):
    if df0['sum'][i].astype(int) > df0['tokens'][i].astype(int):
        ids.append(df0['doc_id'][i])
 
```

```python
print(ids[:10])
```

```python
df0['ratio'] = df0['sum']/df0['tokens']

df0.head()
```

```python
df0.replace([np.inf, -np.inf], np.nan, inplace=True)
df0['ratio'] = df0['ratio'].fillna(0)
df0.head()
```

```python
df0.to_pickle('pkl/new.pkl')
```

```python
plt.figure(figsize = (18,10))
sns.set_style("ticks")

plot = sns.violinplot(x = 'source_mapped', y = 'sum',  scale = "width", data = df0)

plot.set_xlabel("Count", fontsize = 20)
plot.set_ylabel("Source", fontsize = 20)
plot.tick_params(axis='x', labelsize = 15, colors = 'grey')   
plot.tick_params(axis='y', labelsize = 15, colors = 'grey') 
#plot.figure.savefig('count_violin.pdf')
```

```python
fig = sns.catplot(x = 'sum', y = 'source_mapped', kind = 'box', data = df0, height = 10, aspect = 9/7)
#fig.savefig('count_box.pdf')
```

```python
plt.figure(figsize = (18,10))
sns.set_style("ticks")

plot = sns.violinplot(x = 'ratio', y = 'source_mapped',  scale = "width", data = df0)

plot.set_xlabel("Ratio", fontsize = 20)
plot.set_ylabel("Source", fontsize = 20)
plot.tick_params(axis='x', labelsize = 15, colors = 'grey')   
plot.tick_params(axis='y', labelsize = 15, colors = 'grey') 
#plot.figure.savefig('ratio_violin.pdf')
```
